```python
def plot_relation(dic, target_x, target_y):
    x = []
    y = []
    for key in dic:
        # print(key[0])
        if target_x == "sigmoid":
            # x.append(dic[key][target_x]["a"]/dic[key][target_x]["c"])
            x.append(dic[key][target_x]["b"])
        else:
            x.append(dic[key][target_x])
        if target_y == "sigmoid":
            # y.append(dic[key][target_y]["a"]/dic[key][target_y]["c"])
            y.append(dic[key][target_y]["b"])
        else:
            y.append(dic[key][target_y])

        # プロットを作成
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'o', markersize=5, label=f'{target_y} vs {target_x}')
    plt.xlabel(target_x)
    plt.ylabel(target_y)
    plt.title(f'Relation between {target_x} and {target_y}')
    plt.legend()
    plt.grid(True)
    plt.show()
    
        
```

```python
def func(x, a, b, c):
    return a /(np.exp(-b * x) + c)
```

```python
def plot_funcs(params_list, x_range=(0, 10), num_points=1000):
    x = np.linspace(x_range[0], x_range[1], num_points)
    
    plt.figure(figsize=(10, 6))
    for params in params_list:
        a, b, c = params
        y = func(x, a, b, c)
        plt.plot(x, y, label=f'a={a}, b={b}, c={c}, b/c={b/c}, (a/c)*b = {(a/c)*b}, b/(a/c) = {b/(a/c)}')
        # plt.plot(x, y, label=f'a={a}, b={b}, c={c}, b = {b}, a/c={a/c}')
    
    plt.xlabel('x')
    plt.ylabel('func(x)')
    plt.title('Plot of func(x, a, b, c)')
    plt.legend()
    plt.grid(True)
    plt.show()

```

```python
# Example usage
plot_funcs([(2, 1, 1), (2, 1, 2), (2, 2, 1), (2, 1, 0.5)])
```

```python
plot_funcs([(1, 1, 1), (1, 2, 1)]) # bはスピード
```

```python
plot_funcs([(1, 1, 2), (1, 1, 1)]) # cが大きいほど最終値は小さい
```

```python
plot_funcs([(1, 1, 1), (2, 1, 1)]) # aが大きいほど最終値は大きい
```
